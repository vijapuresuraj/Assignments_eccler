```python
import numpy as np
import math
from scipy import stats
```

Given data

```python
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85,
        1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
n = len(data)
mean = np.mean(data)
std_dev = np.std(data, ddof=1)
```

(a) t-distribution

```python
t_value = stats.t.ppf(1 - 0.005, df=n-1)
margin_t = t_value * (std_dev / math.sqrt(n))

ci_t = (mean - margin_t, mean + margin_t)
```

(b) z-distribution

```python
sigma = 0.2
z_value = 2.576
margin_z = z_value * (sigma / math.sqrt(n))

ci_z = (mean - margin_z, mean + margin_z)
```

Output

```python
print("Sample Mean:", mean)
print("Sample Std Dev:", std_dev)

print("\n99% CI using t-distribution:", ci_t)
print("99% CI using z-distribution:", ci_z)
```
