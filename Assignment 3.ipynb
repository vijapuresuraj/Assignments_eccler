#### Exercise 1: Prime Number Check

```python
num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not a Prime Number")
            break
    else:
        print("Prime Number")
else:
    print("Not a Prime Number")
```

#### Exercise 2: Product of Random Numbers

```python
import random
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)
print("Multiply:", num1, "x", num2)

user_answer = int(input("Enter your answer: "))

if user_answer == num1 * num2:
    print("Correct Answer!")
else:
    print("Wrong Answer! Correct answer is:", num1 * num2)
```

#### Exercise 3: Squares of Even Numbers (Chosen Even)

```python
# Printing squares of even numbers from 100 to 200
print("Squares of EVEN numbers from 100 to 200:")

for num in range(100, 201):
    if num % 2 == 0:  # checking even number
        print(num, "->", num ** 2)
```

#### Exercise 4: Word Counter

```python
# Program to count word frequency

input_text = "This is a sample text. This text will be used to demonstrate the word counter."

words = input_text.split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Print result
for word, count in word_count.items():
    print(f"'{word}': {count}")
```

#### Exercise 5: Palindrome Check

```python
# Function to check palindrome

def is_palindrome(text):
    # Remove spaces and convert to lowercase
    cleaned = ''.join(char.lower() for char in text if char.isalnum())
    
    # Check palindrome
    return cleaned == cleaned[::-1]

# Test
input_text = input("Enter a string: ")

if is_palindrome(input_text):
    print("True (Palindrome)")
else:
    print("False (Not Palindrome)")
```
