```python
import numpy as np
from scipy.stats import chi2

#Observed data
observed = np.array([
    [50, 70],
    [80, 100],
    [60, 90],
    [30, 50],
    [20, 50]
])

#Totals
row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
grand_total = observed.sum()

#Expected values
expected = np.outer(row_totals, col_totals) / grand_total

#Chi-square statistic
chi_square = ((observed - expected) ** 2 / expected).sum()

#Degrees of freedom
df = (observed.shape[0] - 1) * (observed.shape[1] - 1)

#Critical value
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, df)

#Output
print("Chi-Square:", chi_square)
print("Degrees of Freedom:", df)
print("Critical Value:", critical_value)

if chi_square > critical_value:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

## Hypothesis Testing

#### 1.State the Hypotheses Statement

Types of Hypothese Sta
Null Hypothesis (H₀):
The weekly operating cost follows the given model
μ=1000+5X

Alternative Hypothesis (H₁):
The weekly operating cost is higher than the model. 
μ>1000+5X

#### 2.Calculate the Test Statistic

Z= (xˉ)-μ/σ/n

xˉ=3050
𝜇=4000
μ=4000
𝜎=5*25=125
n=25


#### 4.Critical Value

Significance level: α = 0.05
Right-tailed test

####  5. Decision

 5. Decision
If Z > 1.645 → Reject H₀
If Z ≤ 1.645 → Fail to Reject H₀

#### 6.Conclusion

There is no strong evidence to support the claim that weekly operating costs are higher than the model.
In fact, the sample mean is much lower than expected.

```python
import math

# Given values
sample_mean = 3050
mu = 1000 + 5 * 600   # theoretical mean
sigma = 5 * 25
n = 25

# Calculate Z statistic
z = (sample_mean - mu) / (sigma / math.sqrt(n))

# Critical value (alpha = 0.05, right-tailed)
z_critical = 1.645

# Output results
print("Theoretical Mean:", mu)
print("Sample Mean:", sample_mean)
print("Standard Deviation:", sigma)
print("Test Statistic (Z):", z)
print("Critical Value:", z_critical)

# Decision
if z > z_critical:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")

# Conclusion
print("Conclusion: No strong evidence that costs are higher than the model.")
```
